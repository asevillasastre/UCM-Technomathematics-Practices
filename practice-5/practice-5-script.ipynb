```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
original = Image.open('no-eq-pic.jpg','r').convert('L')
EQ = np.asarray(original)
```

```python
vect = []
for i in EQ:
    for j in i: 
        vect.append(j/255)
```

```python
def comparativa(bins_):
    plt.hist(vect, density = True, bins= bins_)
    plt.hist(np.arange(0,1,1/bins_), density = True, bins= bins_)
    plt.xlabel("color"); plt.ylabel("probabilidad")
    plt.show()
```

```python
comparativa(50)
```

```python
x_sorted = sorted(vect)
y_sorted = np.arange(0,1,1/len(x_sorted))
plt.plot(y_sorted, x_sorted)
plt.plot(y_sorted, y_sorted)
plt.show()
```

```python
tam = list(original.size)
tam[0] = tam[0] - 1; tam[1] = tam[1] - 1
equalizada = np.zeros(tam)
indexi = 0
for i in EQ:
    indexi += 1
    indexj = 0
    for j in i:
        indexj += 1
        posic = 0
        while j <= x_sorted[posic]:
            posic +=1
        equalizada[indexi+1][indexj+1] = posic
plt.imshow(equalizada)
```
